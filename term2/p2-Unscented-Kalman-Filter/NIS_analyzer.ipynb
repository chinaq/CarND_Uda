# NIS Analyzer

```python
import pandas
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
nis = pandas.read_csv('./build/my_out_file.csv')
nis.sample(3)
```

```python
nis_radar = nis[nis.sensor_type=='R']
nis_radar.sample(5)
nis_radar.plot(title='Radar NIS')
plt.axhline(y=7.815, color='r')
_ = _
```

```python
nis_lidar = nis[nis.sensor_type=='L']
nis_lidar.sample(5)
nis_lidar.plot(title='Lidar NIS')
plt.axhline(y=5.991, color='r')
_ = _
```
